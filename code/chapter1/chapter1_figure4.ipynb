```julia
#This Julia script can be used to reproduce Figure 1.11 in the textbook:
#Emil Bjornson and Ozlem Tugfe Demir (2024),
#%Boston-Delft: Now Publishers, http://dx.doi.org/10.1561/9781638283157
#
#This is version 1.0 (Last edited: 2024-01-17)
#
#License: This code is licensed under the GPLv2 license. If you in any way
#use this code for research that results in publications, please cite our
#textbook as described above. You can find the complete code package at
#https://github.com/emilbjornson/mimobook

using Plots

# Set parameter values

# Considered wavelengths (meters)
λ₁ = 0.3
λ₂ = 0.1
λ₃ = 0.01

# Propagation distance (meters)
d = 1:0.1:100

# Compute channel gain
β₁ₗᵢₙₑₐᵣ = (λ₁/4π)^2 ./d.^2
β₂ₗᵢₙₑₐᵣ = (λ₂/4π)^2 ./d.^2
β₃ₗᵢₙₑₐᵣ = (λ₃/4π)^2 ./d.^2

# Convert to dB
β₁₎ᵦ = 10 .* log10.(β₁ₗᵢₙₑₐᵣ)
β₂₎ᵦ = 10 .* log10.(β₂ₗᵢₙₑₐᵣ)
β₃₎ᵦ = 10 .* log10.(β₃ₗᵢₙₑₐᵣ)

# Plot simulation results
p = plot(d, β₁₎ᵦ, label=raw"$f=1$ GHz", linestyle=:dash, linewidth=2, color=:red)
    plot!(d, β₂₎ᵦ, label=raw"$f=3$ GHz", linestyle=:solid, linewidth=2, color=:blue)
    plot!(d, β₃₎ᵦ, label=raw"$f=30$ GHz", linestyle=:dashdot, linewidth=2, color=:black)

# Customize the plot
plot!(
  label =raw"Latex"
, xlabel = "Distance [m]"
, ylabel = "Channel gain [dB]"
, legend = :topright
, xticks = (0:20:100)
, yticks = (-110:10:-30)
, ylim = (-110, -30)
, grid = :true
, fontsize = 16
)

# Show plot
display(p)
```
